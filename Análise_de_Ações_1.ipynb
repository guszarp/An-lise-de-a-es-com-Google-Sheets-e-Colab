```python
import pandas as pd
import plotly.express as px
```

```python
df_principal = pd.read_excel('/content/acoes_puras.xlsx', sheet_name='Principal')
df_principal.head(10)
```

```python
df_total_acoes = pd.read_excel('/content/acoes_puras.xlsx', sheet_name='Total_de_acoes')
df_total_acoes.head(10)
```

```python
df_ticker = pd.read_excel('/content/acoes_puras.xlsx', sheet_name='Ticker')
df_ticker.head(10)
```

```python
df_gpt = pd.read_excel('/content/acoes_puras.xlsx', sheet_name='Chatgpt')
df_gpt.head(10)
```

```python
df_principal = df_principal[['Ativo','Data','Último (R$)','Var. Dia (%)']].copy()
df_principal
```

```python
df_principal = df_principal.rename(columns={'Último (R$)':'valor final', 'Var. Dia (%)':'variacao_dia_pct'}).copy()
df_principal
```

```python
df_principal = df_principal.rename(columns={'Ativo':'ativo', 'Data':'data'}).copy()
df_principal
```

```python
df_principal['var_pct'] = df_principal['variacao_dia_pct'] / 100
df_principal['valor_inicial'] = df_principal['valor final'] / (df_principal['var_pct'] + 1)

df_principal
```

```python
df_principal = df_principal.rename(columns={'valor final':'valor_final'}).copy()
df_principal
```

```python
df_principal = df_principal.merge(df_total_acoes, left_on = 'ativo', right_on = 'Código', how = 'left')
df_principal
```

```python
df_principal = df_principal.drop(columns=['Código'])
df_principal
```

```python
df_principal['var_volume'] = (df_principal['valor_final'] - df_principal['valor_inicial']) * df_principal['Qtde. Teórica']
df_principal
```

```python
pd.options.display.float_format = '{:.2f}'.format
df_principal
```

```python
df_principal['Qtde. Teórica'] = df_principal['Qtde. Teórica'].astype(int)
df_principal
```

```python
df_principal = df_principal.rename(columns={'Qtde. Teórica':'qtd_teorica'}).copy()
df_principal
```

```python
df_principal['resultado'] = df_principal['var_volume'].apply(lambda x: 'Subiu' if x > 0 else ('Desceu' if x < 0 else 'Não variou'))
df_principal
```

```python
df_principal = df_principal.merge(df_ticker, left_on = 'ativo', right_on = 'Ticker', how = 'left')
df_principal = df_principal.drop(columns=['Ticker'])
df_principal
```

```python
df_principal = df_principal.rename(columns={'Nome':'nome'}).copy()
df_principal
```

```python
df_principal = df_principal.merge(df_gpt, left_on = 'nome', right_on = 'Nome da empresa', how = 'left')
df_principal = df_principal.drop(columns=['Nome da empresa'])
df_principal = df_principal.rename(columns={'Segmento':'segmento','Idade em anos':'idade_anos'}).copy()
df_principal
```

```python
df_principal['faixa_idade'] = df_principal['idade_anos'].apply(lambda x: 'Mais que 100 anos' if x > 100 else ('Menor que 50 anos' if x < 50 else 'Entre 50 e 100 anos'))
df_principal
```

```python
df_principal = df_principal.drop_duplicates(subset='ativo')
df_principal
```

```python
maior_var = df_principal['var_volume'].max()
menor_var = df_principal['var_volume'].min()
media_var = df_principal['var_volume'].mean()

media_var_subiu = df_principal[df_principal['resultado'] == 'Subiu']['var_volume'].mean()
media_var_desceu = df_principal[df_principal['resultado'] == 'Desceu']['var_volume'].mean()

print(f'Maior Valor de Variação R$ {maior_var:.2f}')
print(f'Menor Valor de Variação R$ {menor_var:.2f}')
print(f'Média de Valor de Variação R$ {media_var:.2f}')
print(f'Média Valor de Variação que subiu R$ {media_var_subiu:.2f}')
print(f'Média Valor de Variação  que desceu R$ {media_var_desceu:.2f}')
```

```python
df_subiu = df_principal[df_principal['resultado'] == 'Subiu']
df_subiu
```

```python
df_analise_segmento_subiu = df_subiu.groupby('segmento')['var_volume'].sum().reset_index()
df_analise_segmento_subiu
```

```python
df_analise_saldo = df_principal.groupby('resultado')['var_volume'].sum().reset_index()
df_analise_saldo
```

```python
df_analise_idade = df_principal.groupby('faixa_idade')['var_volume'].sum().reset_index()
df_analise_idade
```

```python
fig_1 = px.bar(df_analise_saldo, x = 'resultado', y = 'var_volume', text = 'var_volume', title = 'Variação em Reais X Resultado')
fig_1.update_traces(textposition='outside')
fig_1.update_layout(yaxis_tickformat = ',.2f')
fig_1.show()
```

```python
fig_2 = px.pie(df_analise_segmento_subiu, values='var_volume', names='segmento', title='Variação em Reais X Segmento que Subiu')
fig_2.show()
```

```python
df_analise_idade['qtd_empresas'] = df_principal.groupby('faixa_idade')['ativo'].count().reset_index()['ativo']
df_analise_idade
```

```python
fig_3 = px.bar(df_analise_idade, x = 'faixa_idade', y = 'var_volume', text = 'var_volume')
fig_3.update_traces(textposition='outside')
fig_3.update_layout(yaxis_tickformat = ',.2f')
fig_3.show()
```

```python
fig_4 = px.scatter(df_principal, x = 'valor_inicial', y = 'valor_final', title = 'Dispersão entre Valor Incial X Valor Final')
fig_4.show()
```
