```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('database.csv')
```

```python
df.head()
```

```python
# Converter a coluna Launch Date para Datetime

df['Launch Date'] = pd.to_datetime(df['Launch Date'])
df['Year'] = df['Launch Date'].dt.year
df.head()
```

```python
# Contagem de lançamentos por ano

launches_per_year = df['Year'].value_counts().sort_index()
df.head()
```

```python
# Gráfico de série temporal

plt.figure(figsize = (10, 6))
launches_per_year.plot(kind = 'line', marker = 'o', title = 'SpaceX Launches Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Launches')
plt.grid(True)
plt.show
```

```python
import warnings
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA

warnings.filterwarnings("ignore")
```

```python
# Modelando Arima para a previsão

model = ARIMA(launches_per_year, order=(5, 1, 0))
model_fit = model.fit()
forecast = model_fit.forecast(steps = 1)
forecast_value = forecast.iloc[0]
print(f'Previsão de lançamentos para 2023: {forecast_value}')
```

```python
# Resultados:

# A tendencia é que haja uma crescente no número de lançamentos para os próximos anos
# Não ha sazonalidades
# Há possibilidade de variações nos dados por conta de fatores aleatórios, como falhas e condições meteorológicas
```
